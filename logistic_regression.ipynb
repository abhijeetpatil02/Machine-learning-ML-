Logistic regression
predict pass or fail based on number of hours study

```python
from sklearn. linear_model import LogisticRegression

```

```python
x=[[1],[2],[3],[4],[5]]
y=[0,0,1,1,1]

model=LogisticRegression()

model.fit(x,y)

hour=float(input("enter how many hour study"))

result=model.predict([[hour]])[0]

if result==1:
    print(f"based on hours{hour},your likely to PASS")
else:
    print(f"based on hours{hour},your likely to FAIL")
```
