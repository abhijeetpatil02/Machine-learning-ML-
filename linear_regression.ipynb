```python
from sklearn.linear_model import LinearRegression
```

```python
x=[[1],[2],[3],[4],[5]]
y=[40,50,65,75,90]

model=LinearRegression()

model.fit(x,y)

hour=float(input("enter how many hour study"))

predicted_marks=model.predict([[hour]])

print(predicted_marks)
```
